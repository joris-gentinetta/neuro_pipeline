```python
experiment_name = '2021-02-26_mBWfus012_OF_ephys'
```

```python
import numpy as np
import pickle5 as pkl
import pandas as pd
import matplotlib.pyplot as plt
import os
from scipy.ndimage.filters import gaussian_filter1d
from scipy.ndimage.filters import gaussian_filter
import math
from matplotlib import cm
import shutil
```

```python
if experiment_name[-7] == 'M':
    environment = 'EZM'
elif experiment_name[-7] == 'a':
    environment = 'arena'
elif experiment_name[-7] == 'F':
    environment = 'OFT'
else:
    raise NameError('environment not supported')
```

```python
sorter = 'circus'
movefile = r'E:\behavioral_results/' + environment + '_movement/' + experiment_name + '_movement.pkl'
data_folder = r'E:\ephys_data/'
folder = data_folder + experiment_name +'/' + sorter + '/'
datafile = folder + 'pp_data.npy'
ptriggerfile = folder + 'trigger.npy'
mtriggerfile = data_folder +'Anxiety_video_triggers.xlsx'
plot_folder = folder + 'plots/'
target_folder = r'E:/ephys_data/' + experiment_name +'/circus/'
```

```python
if(os.path.exists(plot_folder)):
    print('overwrite ', plot_folder, '? [y/n]')
    if input() == 'y':
        shutil.rmtree(plot_folder, ignore_errors=True)
        os.mkdir(plot_folder)
else:
    os.mkdir(plot_folder)
```

```python
physio_trigger = int(np.load(ptriggerfile)*50//20000)
video_trigger = int(pd.read_excel(mtriggerfile, index_col = 'session').loc[experiment_name, 'trigger/s']*50)
```

```python
data = np.load(datafile) 
data = data.astype(np.float32, copy = False)
```

```python
sigma = 100
for i in range(data.shape[0]):
    data[i] = gaussian_filter1d(data[i], sigma=sigma)
```

```python
with open(movefile, 'rb') as f:
    movement = pkl.load(f)
```

```python
print(movement['calib_traj_y'].shape[0])
print(data.shape[1])
```

```python
# offset for nan in 2021-02-26_mBWfus012_EZM_ephys_movement
nans = np.where(np.isnan(movement['calib_traj_y'][video_trigger:]))[0]
if nans.shape[0] == 0:
    off = 0
else:
    off = nans[-1] + 1
    
cluster_names = np.load(folder + 'cluster_names.npy')
```

```python
print(off)
```

```python
n = 150 #number of steps in grid # 1/binsize standard: n=150

grid = np.zeros((n,n,data.shape[0]), dtype = np.float32)
xyv = np.empty((3, min(len(movement['calib_traj_x'].index)-video_trigger - off, data.shape[1]-physio_trigger - off)))

#commmand to start the viewer
print('circus-gui-python ' + target_folder + 'dat_files/' + experiment_name + '_0.dat')

for i in range(data.shape[0]):
    xyv[0] = movement['calib_traj_x'][video_trigger + off : xyv.shape[1] + video_trigger + off]
    xyv[1] = movement['calib_traj_y'][video_trigger + off : xyv.shape[1] + video_trigger + off]
    xyv[2] = data[i][physio_trigger + off : xyv.shape[1] + physio_trigger + off]
    xyv[0] -= xyv[0].min()
    xyv[1] -= xyv[1].min()
    sx = xyv[0].max()
    sy = xyv[1].max()
    

    for x in range(n):
        for y in range(n):
            boolean = np.logical_and(np.logical_and(xyv[0] > x/n*sx, xyv[0] < (x+1)/n*sx), np.logical_and(xyv[1] > y/n*sy, xyv[1] < (y+1)/n*sy))
            bsum = np.sum(boolean)
            if bsum != 0:
                grid[x,y,i] = np.sum(xyv[2][boolean])/bsum*100

    fig = plt.figure(figsize=(5,5))
    plt.imshow(grid[:,:,i], 'jet')
    plt.savefig(plot_folder + experiment_name + '_gaussian1d' + str(sigma) + '_n' + str(n) + '_' + str(cluster_names[i]) + '.jpg')
    if i != 0:
        plt.title('firing rate unit '+ str(cluster_names[i]))
    else:
        plt.title('firing rate all mua')
    plt.show()
fig = plt.figure(figsize=(5,5))
plt.imshow(np.sum(grid[:,:,1:], axis = 2), 'jet')
plt.savefig(plot_folder + experiment_name + '_gaussian1d' + str(sigma) + '_n' + str(n) + '_' + 'all_units' + '.jpg')
plt.title('firing rate all units')
plt.show()
```

```python
print(data.mean(axis=1)*50)
```

```python
#circle plot
n = 360 #number of steps in circle standard n=360
sigma2 = 1
grid = np.zeros((data.shape[0], n), dtype = np.float32)
xyv = np.empty((4, min(len(movement['calib_traj_x'].index)-video_trigger - off, data.shape[1]-physio_trigger - off)))

for i in range(data.shape[0]):
    xyv[0] = movement['calib_traj_x'][video_trigger + off : xyv.shape[1] + video_trigger + off]
    xyv[1] = movement['calib_traj_y'][video_trigger + off : xyv.shape[1] + video_trigger + off]
    xyv[2] = data[i][physio_trigger + off : xyv.shape[1] + physio_trigger + off]
    xyv[0] -= xyv[0].min()
    xyv[1] -= xyv[1].min()
    middle_x = xyv[0].max()/2
    middle_y = xyv[1].max()/2
    xyv[0] -= middle_x
    xyv[1] -= middle_y
    xyv[3] = np.arctan2(xyv[0], xyv[1])
    xyv[3] += (xyv[3]<0)*2*math.pi
    for r in range(n):
        boolean = np.logical_and(xyv[3] > r*math.pi*2/n, xyv[3] < (r+1)*math.pi*2/n)
        bsum = np.sum(boolean)
        if bsum != 0:
            grid[i,r] = np.sum(xyv[2][boolean])/np.sum(boolean)*100
    grid[i] = gaussian_filter1d(grid[i], sigma=sigma2, mode='wrap')
    fig = plt.figure(figsize=(5,5))
    plt.pie(np.ones(n),colors = cm.jet(plt.Normalize()(grid[i])))
    my_circle=plt.Circle( (0,0), 0.8, color='white')
    p=plt.gcf()
    p.gca().add_artist(my_circle)
    plt.savefig(plot_folder + experiment_name + '_pie_gaussian1d' + str(sigma) + '_2gaussian1d' + str(sigma2) + '_n' + str(n) + '_' + str(i) + '.jpg')
    if i != 0:
        plt.title('firing rate unit '+ str(cluster_names[i]))
    else:
        plt.title('firing rate all mua')
    plt.show()
fig = plt.figure(figsize=(5,5))
plt.pie(np.ones(n),colors = cm.jet(plt.Normalize()(np.sum(grid, axis=0))))
my_circle=plt.Circle( (0,0), 0.8, color='white')
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.savefig(plot_folder + experiment_name + '_pie_gaussian1d' + str(sigma) + '_2gaussian1d' + str(sigma2) + '_n' + str(n) + '_' + 'all_units' + '.jpg')
plt.title('firing rate all units')
plt.show()
```
