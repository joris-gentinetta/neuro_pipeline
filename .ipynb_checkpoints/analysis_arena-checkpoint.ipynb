```python
import numpy as np
import pickle5 as pkl
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datafile = r'E:\ephys_data\2021-02-26_mBWfus012_arena_ephys\circus/pp_data.npy'
movefile = r'E:\behavioral_results\arena_movement/2021-02-26_mBWfus012_arena_ephys_movement.pkl'
```

```python
data = np.load(datafile) 
```

```python
with open(movefile, 'rb') as f:
    movement = pkl.load(f)
```

```python
for i in range(data.shape[0]):
    #i=1 #cluster number
    n=200 #number of steps in grid # 1/binsize

    xyv = np.empty((3, min(len(movement['calib_traj_x'].index), data.shape[1])))
    xyv[0] = movement['calib_traj_x'][:xyv.shape[1]]
    xyv[1] = movement['calib_traj_y'][:xyv.shape[1]]
    xyv[2] = data[i][:xyv.shape[1]]
    xyv[0] -= xyv[0].min()
    xyv[1] -= xyv[1].min()
    sx = xyv[0].max()
    sy = xyv[1].max()
    grid = np.zeros((n,n), dtype = np.uint8)

    for x in range(n):
        for y in range(n):
            boolean = np.logical_and(np.logical_and(xyv[0] > x/n*sx, xyv[0] < (x+1)/n*sx), np.logical_and(xyv[1] > y/n*sy, xyv[1] < (y+1)/n*sy))
            bsum = np.sum(boolean)
            if bsum != 0:
                grid[x,y] = np.sum(xyv[2][boolean])/np.sum(boolean)*10

    fig = plt.figure(figsize=(10,10))
    plt.imshow(grid, "jet")
    plt.title("spikeintensity "+ str(i))
    plt.show()
```
