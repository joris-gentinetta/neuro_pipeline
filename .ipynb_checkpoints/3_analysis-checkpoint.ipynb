```python

def plot_classic(sigma = 10, minp = 0, maxp = 95, n = 150):
    mode = 'classic'
    make_path_visible = 0.0001
    file_name = plot_folder + experiment_name + '_' + mode + '_sigma' + str(sigma) + '_n' + str(n) + '_minp' + str(minp) + '_maxp' + str(maxp) + '_'
    data = np.empty(raw_data.shape, dtype = np.float32) 


    for i in range(raw_data.shape[0]):
        data[i] = gaussian_filter1d(raw_data[i], sigma=sigma)
    grid = np.zeros((n,n,data.shape[0]), dtype = np.float32)
    xyv = np.empty((3, min(len(movement['calib_traj_x'].index)-video_trigger - off, data.shape[1]-physio_trigger - off)), dtype = np.float32)

    cmap = copy.copy(mpl.cm.get_cmap('jet'))
    cmap.set_bad(color='grey')
    for i in range(data.shape[0]):
        xyv[0] = movement['calib_traj_x'][video_trigger + off : xyv.shape[1] + video_trigger + off]
        xyv[1] = movement['calib_traj_y'][video_trigger + off : xyv.shape[1] + video_trigger + off]
        xyv[2] = data[i][physio_trigger + off : xyv.shape[1] + physio_trigger + off] + make_path_visible
        xyv[0] -= xyv[0].min()
        xyv[1] -= xyv[1].min()
        sx = xyv[0].max()
        sy = xyv[1].max()

        for x in range(n):
            for y in range(n):
                boolean = np.logical_and(np.logical_and(xyv[0] > x/n*sx, xyv[0] < (x+1)/n*sx), np.logical_and(xyv[1] > y/n*sy, xyv[1] < (y+1)/n*sy))
                bsum = np.sum(boolean)
                if bsum != 0:
                    grid[x,y,i] = np.sum(xyv[2][boolean])/bsum

        vmin = np.percentile(grid[:,:,i][np.where(grid[:,:,i]>0)], minp)
        vmax = np.percentile(grid[:,:,i][np.where(grid[:,:,i]>0)], maxp)

        fig = plt.figure(figsize=(5,5))
        

        im = plt.imshow(np.ma.masked_where(grid[:,:,i]==0, grid[:,:,i]).T , cmap=cmap, origin='upper', interpolation = 'none', vmin=vmin, vmax=vmax)
        plt.colorbar(im,fraction=0.046, pad=0.04)
        plt.savefig(file_name + str(cluster_names[i]) +'.jpg')
        if i != 0:
            plt.title('firing rate unit '+ str(cluster_names[i]))
        else:
            plt.title('firing rate all mua')
        #plt.show()
        plt.close(fig)

    unit_sum = np.sum(grid[:,:,1:], axis = 2)
    vmin = np.percentile(unit_sum[np.where(unit_sum>0)], minp)
    vmax = np.percentile(unit_sum[np.where(unit_sum>0)], maxp)

    fig = plt.figure(figsize=(5,5))
    im = plt.imshow(np.ma.masked_where(unit_sum==0, unit_sum).T , cmap=cmap, origin='upper', interpolation = 'none', vmin=vmin, vmax=vmax)
    plt.colorbar(im, fraction=0.046, pad=0.04)
    plt.savefig(file_name + 'all_units' + '.jpg')
    plt.title('firing rate all units')
    #plt.show()
    plt.close(fig)

def plot_masked(minp = 0, maxp = 90, n = 150):
    mode = 'masked'
    make_path_visible = 0.0001
    file_name = plot_folder + experiment_name + '_' + mode + '_n' + str(n) + '_minp' + str(minp) + '_maxp' + str(maxp) + '_'

    data = raw_data
    grid = np.zeros((n,n,data.shape[0]), dtype = np.float32)
    xyv = np.empty((3, min(len(movement['calib_traj_x'].index)-video_trigger - off, data.shape[1]-physio_trigger - off)), dtype = np.float32)

    cmap = copy.copy(mpl.cm.get_cmap("Blues"))
    cmap.set_bad(color='grey')
    for i in range(data.shape[0]):
        xyv[0] = movement['calib_traj_x'][video_trigger + off : xyv.shape[1] + video_trigger + off]
        xyv[1] = movement['calib_traj_y'][video_trigger + off : xyv.shape[1] + video_trigger + off]
        xyv[2] = data[i][physio_trigger + off : xyv.shape[1] + physio_trigger + off] + make_path_visible
        xyv[0] -= xyv[0].min()
        xyv[1] -= xyv[1].min()
        sx = xyv[0].max()
        sy = xyv[1].max() 

        for x in range(n):
            for y in range(n):
                boolean = np.logical_and(np.logical_and(xyv[0] > x/n*sx, xyv[0] < (x+1)/n*sx), np.logical_and(xyv[1] > y/n*sy, xyv[1] < (y+1)/n*sy))
                bsum = np.sum(boolean)
                if bsum != 0:
                    grid[x,y,i] = np.sum(xyv[2][boolean])/bsum

        vmin = np.percentile(grid[:,:,i][np.where(grid[:,:,i]>0)], minp)
        vmax = np.percentile(grid[:,:,i][np.where(grid[:,:,i]>0)], maxp)

        fig = plt.figure(figsize=(5,5))
        im = plt.imshow(np.ma.masked_where(grid[:,:,i]==0, grid[:,:,i]).T , cmap=cmap, origin='upper', interpolation = 'none', vmin=vmin, vmax=vmax)
        plt.colorbar(im,fraction=0.046, pad=0.04)
        plt.savefig(file_name + str(cluster_names[i]) +'.jpg')

        if i != 0:
            plt.title('firing rate unit '+ str(cluster_names[i]))
        else:
            plt.title('firing rate all mua')
        #plt.show()
        plt.close(fig)

    unit_sum = np.sum(grid[:,:,1:], axis = 2)
    vmin = np.percentile(unit_sum[np.where(unit_sum>0)], minp)
    vmax = np.percentile(unit_sum[np.where(unit_sum>0)], maxp)

    fig = plt.figure(figsize=(5,5))
    im = plt.imshow(np.ma.masked_where(unit_sum==0, unit_sum).T , cmap=cmap, origin='upper', interpolation = 'none', vmin=vmin, vmax=vmax)
    plt.colorbar(im,fraction=0.046, pad=0.04)
    plt.savefig(file_name + 'all_units' + '.jpg')
    plt.title('firing rate all units')
    #plt.show()
    plt.close(fig)
    return

def plot_circle(n = 360, sigma = -1):  
    mode = 'circle'
    if sigma<0:
        sigma = n/8/4
    file_name = plot_folder + experiment_name + '_' + mode + '_sigma' + str(sigma) + '_n' + str(n) + '_'
  

    data = raw_data
    grid = np.zeros((data.shape[0], n), dtype = np.float32)
    xyv = np.empty((4, min(len(movement['calib_traj_x'].index)-video_trigger - off, data.shape[1]-physio_trigger - off)))

    for i in range(data.shape[0]):
        xyv[0] = movement['calib_traj_x'][video_trigger + off : xyv.shape[1] + video_trigger + off]
        xyv[1] = movement['calib_traj_y'][video_trigger + off : xyv.shape[1] + video_trigger + off]
        xyv[2] = data[i][physio_trigger + off : xyv.shape[1] + physio_trigger + off]
        xyv[0] -= xyv[0].min()
        xyv[1] -= xyv[1].min()
        middle_x = xyv[0].max()/2
        middle_y = xyv[1].max()/2
        xyv[0] -= middle_x
        xyv[1] -= middle_y 
        xyv[1] *= -1
        xyv[3] = np.arctan2(xyv[1], xyv[0])
        xyv[3] += (xyv[3]<0)*2*math.pi
        for r in range(n):
            boolean = np.logical_and(xyv[3] > r*math.pi*2/n, xyv[3] < (r+1)*math.pi*2/n)
            bsum = np.sum(boolean)
            if bsum != 0:
                grid[i,r] = np.sum(xyv[2][boolean])/np.sum(boolean)
        grid[i] = gaussian_filter1d(grid[i], sigma=sigma, mode='wrap')
        fig = plt.figure(figsize=(5,5))
        plt.pie(np.ones(n),colors = cm.jet(plt.Normalize()(grid[i])))
        plt.colorbar(cm.ScalarMappable(norm=mpl.colors.Normalize(vmin=grid[i].min(), vmax=grid[i].max()), cmap='jet'), shrink = 0.6)
        my_circle=plt.Circle( (0,0), 0.8, color='white')
        fig.gca().add_artist(my_circle)
        plt.savefig(file_name + str(cluster_names[i]) +'.jpg')
        if i != 0:
            plt.title('firing rate unit '+ str(cluster_names[i]))
        else:
            plt.title('firing rate all mua')
        #plt.show()
        plt.close(fig)

    unit_sum = np.sum(grid[1:], axis=0)
    fig = plt.figure(figsize=(5,5))
    fig.colorbar(cm.ScalarMappable(norm=mpl.colors.Normalize(vmin=unit_sum.min(), vmax=unit_sum.max()), cmap='jet'), shrink=0.6)
    plt.pie(np.ones(n),colors = cm.jet(plt.Normalize()(unit_sum)))
    my_circle=plt.Circle( (0,0), 0.8, color='white')
    fig.gca().add_artist(my_circle)
    plt.savefig(file_name + 'all_units' + '.jpg')
    plt.title('firing rate all units')
    #plt.show()
    plt.close(fig)
    return

```

```python
def plot_grid(minp = 0, maxp = 100, n = 5):
    mode = 'jet'
    make_path_visible = 0.0001
    file_name = plot_folder + experiment_name + '_' + mode + '_n' + str(n) + '_minp' + str(minp) + '_maxp' + str(maxp) + '_'

    data = raw_data  
    grid = np.zeros((n,n,data.shape[0]), dtype = np.float32)
    
    xyv = np.empty((3, min(len(movement['calib_traj_x'].index)-video_trigger - off, data.shape[1]-physio_trigger - off)), dtype = np.float32)

    cmap = copy.copy(mpl.cm.get_cmap("Blues"))
    cmap.set_bad(color='grey')
    for i in range(data.shape[0]):
        xyv[0] = movement['calib_traj_x'][video_trigger + off : xyv.shape[1] + video_trigger + off]
        xyv[1] = movement['calib_traj_y'][video_trigger + off : xyv.shape[1] + video_trigger + off]
        xyv[2] = data[i][physio_trigger + off : xyv.shape[1] + physio_trigger + off] + make_path_visible
        xyv[0] -= xyv[0].min()
        xyv[1] -= xyv[1].min()
        sx = xyv[0].max()
        sy = xyv[1].max() 

        for x in range(n):
            for y in range(n):
                boolean = np.logical_and(np.logical_and(xyv[0] > x/n*sx, xyv[0] < (x+1)/n*sx), np.logical_and(xyv[1] > y/n*sy, xyv[1] < (y+1)/n*sy))
                bsum = np.sum(boolean)
                if bsum != 0:
                    grid[x,y,i] = np.sum(xyv[2][boolean])/bsum
                    
        for d in range(n):
            grid[1:n-1, d, i] = np.mean(grid[1:n-1, d, i])
        
        for d in range(n):
            grid[d, 1:n-1, i] = np.mean(grid[d, 1:n-1, i])

        vmin = np.percentile(grid[:,:,i][np.where(grid[:,:,i]>0)], minp)
        vmax = np.percentile(grid[:,:,i][np.where(grid[:,:,i]>0)], maxp)

        fig = plt.figure(figsize=(5,5))
        im = plt.imshow(np.ma.masked_where(grid[:,:,i]==0, grid[:,:,i]).T , cmap=cmap, origin='upper', interpolation = 'none', vmin=vmin, vmax=vmax)
        plt.colorbar(im,fraction=0.046, pad=0.04)
        plt.savefig(file_name + str(cluster_names[i]) +'.jpg')

        if i != 0:
            plt.title('firing rate unit '+ str(cluster_names[i]))
        else:
            plt.title('firing rate all mua')
        #plt.show()
        plt.close(fig)

    unit_sum = np.sum(grid[:,:,1:], axis = 2)
    vmin = np.percentile(unit_sum[np.where(unit_sum>0)], minp)
    vmax = np.percentile(unit_sum[np.where(unit_sum>0)], maxp)

    fig = plt.figure(figsize=(5,5))
    im = plt.imshow(np.ma.masked_where(unit_sum==0, unit_sum).T , cmap=cmap, origin='upper', interpolation = 'none', vmin=vmin, vmax=vmax)
    plt.colorbar(im,fraction=0.046, pad=0.04)
    plt.savefig(file_name + 'all_units' + '.jpg')
    plt.title('firing rate all units')
    #plt.show()
    plt.close(fig)
    return
```

```python
for experiment_name in experiment_names:
    if experiment_name[-7] == 'M':
        environment = 'EZM'
    elif experiment_name[-7] == 'a':
        environment = 'arena'
    elif experiment_name[-7] == 'F':
        environment = 'OFT'
    else:
        raise NameError('environment not supported')
    #movefile = r'E:\behavioral_results/' + environment + '_movement/' + experiment_name + '_movement.pkl'
    movefile = data_folder + animal + '/' + experiment_name + '/ephys_processed/' + experiment_name + '_movement.pkl'
    datafile = target_folder + experiment_name + '.npy'
    ptriggerfile = target_folder + experiment_name + '_trigger.npy'
    mtriggerfile = data_folder + animal + '/' + experiment_name + '/log.txt'
    plot_folder = all_plots + experiment_name + '/'

    if(os.path.exists(plot_folder)):
        print('overwrite ', plot_folder, '? [y/n]')
        if input() == 'y':
            shutil.rmtree(plot_folder, ignore_errors=True)
            os.mkdir(plot_folder)
    else:
        os.mkdir(plot_folder)

    with open(mtriggerfile, 'r') as file:
        data = file.read().replace('\n', ' ')
    point = data.index('.')
    video_trigger = int(float(data[point-2:point+3])*framerate)

    physio_trigger = int(np.load(ptriggerfile)*framerate//20000)

    raw_data = np.load(datafile) * framerate
    raw_data = raw_data.astype(np.float32, copy = False)

    with open(movefile, 'rb') as f:
        movement = pkl.load(f)

    # offset for nan in 2021-02-26_mBWfus012_EZM_ephys_movement
    nans = np.where(np.isnan(movement['calib_traj_y'][video_trigger:]))[0]
    if nans.shape[0] == 0:
        off = 0
    else:
        off = nans[-1] + 1

    cluster_names = np.load(target_folder + 'cluster_names.npy')
#################################
    plot_classic()
#################################
    plot_masked()
#################################
    if environment == 'EZM':
        plot_circle()
    else:
        plot_grid()
```
