```python
## only cell to adapt
experiment_name = '2021-02-19_mBWfus009_OF_ephys'
max_impedance = 2000000 
```

```python
import numpy as np
from load_intan_rhd_format.load_intan_rhd_format import read_data
import os
import shutil
from IPython.utils import io
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
#folder preparation
data_folder = r'E:/ephys_data/' + experiment_name +'/'
sub_folder = 'circus'
rhd_folder = data_folder + 'ephys/'
target_folder = data_folder + sub_folder + '/'
templates = r'E:/pipe/spykingcircus_sorter/templates/'
rhd_files = os.listdir(rhd_folder)
param_file = experiment_name + '_0.params'
```

```python
#folder creation
if(os.path.exists(target_folder)):
    print('overwrite ', target_folder, '? [y/n]')
    if input() == 'y':
        shutil.rmtree(target_folder, ignore_errors=True)
        os.mkdir(target_folder)
        os.mkdir(target_folder + 'dat_files')
        os.mkdir(target_folder + 'dat_files_mod')
else:
    os.mkdir(target_folder)
    os.mkdir(target_folder + 'dat_files')
    os.mkdir(target_folder + 'dat_files_mod')
```

```python
#create trigger file
with io.capture_output():
    r = read_data(rhd_folder+ rhd_files[0])
trigger = np.array(r['board_dig_in_data'][0])
i, = np.where(trigger==1)
np.save(target_folder+'trigger', i[0])
```

```python
%%capture  

amp_channels = read_data(rhd_folder+ rhd_files[0])['amplifier_channels']

#pta[x] denotes the channel of pad x+1
pta = [28, 29, 27, 12, 20, 21, 11, 7, 1, 8, 10, 15, 18, 23, 26, 31, 2, 6, 9, 14, 17, 22, 25, 30, 19, 24, 0, 13, 16, 5, 4, 3, 32, 45, 43, 56, 35, 40, 60, 52, 33, 38, 41, 46, 49, 54, 57, 62, 34, 39, 42, 47, 50, 55, 58, 63, 44, 36, 59, 61, 37, 48, 53, 51]

#pad[x] denotes the pad corresponding to data['amplifier_data'][x]
pad = [pta[int(channel['native_channel_name'][2:])] + 1 for channel in amp_channels]

# probe file 'graph'
graph = []  

# probe file 'geometry'
geometry={}
for i in range(len(pad)):
    if pad[i] < 33:
        geometry[i] = (0, (pad[i] - 32) * 70)
    else:
        geometry[i] = (3000, (pad[i] - 64) * 70)
# param file 'nchannels'        
total_nb_channels = len(amp_channels)
```

```python
#create .dat files
#time: 02:28min for 25 files
for index, rhdfile in tqdm(enumerate(rhd_files)):
    with io.capture_output():
        x = np.transpose(np.array(read_data(rhd_folder + rhdfile)['amplifier_data'], dtype = np.int16));
    x.tofile(target_folder + 'dat_files/' + experiment_name + '_' + str(index) + '.dat')
```

```python
#prepare data for plotting
with io.capture_output():
    x = np.array(read_data(rhd_folder + rhd_files[0])['amplifier_data'], dtype = np.int16);
mPFC = np.zeros(x.shape[1], dtype = np.int16)
n_mPFC = 0
vH = np.zeros(x.shape[1], dtype = np.int16)
n_vH = 0
for i in range(x.shape[0]):
    if pad[i] < 33: 
        mPFC += x[i]
        n_mPFC += 1
    else: 
        vH += x[i]
        n_vH += 1
for i in range(x.shape[0]):
    if pad[i] < 33:
        x[i] = (x[i] - mPFC/n_mPFC).astype(np.int16)
            
    else: 
        x[i] = (x[i] - vH/n_vH).astype(np.int16)       
```

```python
#plot channels
lw =0.4
d=x[:,x.shape[1]//10*8:x.shape[1]//10*9]
zero = np.zeros(d.shape[1])
frontier = [1 if i % 2 == 0 else 0 for i in range(d.shape[1]) ]
fig = plt.figure(figsize = (20,30))
gs = fig.add_gridspec(64, hspace=0)
axs = gs.subplots(sharex=True, sharey=True)
for i, ax in tqdm(enumerate(axs)):
    for ind in range(64):
        if ind >= len(pad):
            if i == 31:
                ax.plot(zero, label = '0, frontier', linewidth = lw)
            else:
                ax.plot(zero, label = '0', linewidth = lw)
        elif pad[ind] == i+1:
            if i == 31:
                if amp_channels[ind]['electrode_impedance_magnitude'] > max_impedance:
                    ax.plot(d[ind], label =str(ind) + ', impedance' + ', frontier', linewidth = lw)
                else:
                    ax.plot(d[ind], label = str(ind) + ', frontier', linewidth = lw)
            else:
                if amp_channels[ind]['electrode_impedance_magnitude'] > max_impedance:
                    ax.plot(d[ind], label = str(ind) + ', impedance', linewidth = lw)
                else:
                    ax.plot(d[ind], label = str(ind), linewidth = lw)
    ax.legend(loc = 'upper right')
plt.ylim(-150, 150)
plt.show()
```

```python
#probe file 'channels'
discarded_channels = np.array(input('discarded channels seperated by space: ').split(), dtype = np.int16)
channels = [i for i, channel in enumerate(amp_channels) if channel['electrode_impedance_magnitude'] < max_impedance and i not in discarded_channels]
np.save(target_folder + 'discarded_channels', discarded_channels)
```

```python
%%capture cap1 --no-stderr
#prepare probe.prb file
radius = 100
print('total_nb_channels = ', total_nb_channels)
print('radius = ', radius)
print('graph = ', graph)
print('channels = ', channels)
print('geometry = ', geometry)                       
```

```python
#create probe.prb file
with open(target_folder + 'probe.prb', 'w') as f:
    f.write(cap1.stdout)   
with open(templates+'probespecs.prb', 'r') as f:
    t=f.read()   
with open(target_folder + 'probe.prb', 'a') as f:
    f.write(t) 
```

```python
%%capture cap2 --no-stderr
#prepare parameters.params file
print('[data]')
print('nb_channels = ', total_nb_channels)
print('mapping = ', target_folder + 'probe.prb')
print('output_dir = ', target_folder + 'dat_files_mod')
```

```python
#create parameters.params file
with open(target_folder + 'dat_files/' + param_file, 'w') as f:
    f.write(cap2.stdout)   
with open(templates+'parameters.params', 'r') as f:
    t=f.read()   
with open(target_folder + 'dat_files/' + param_file, 'a') as f:
    f.write(t)  
```
